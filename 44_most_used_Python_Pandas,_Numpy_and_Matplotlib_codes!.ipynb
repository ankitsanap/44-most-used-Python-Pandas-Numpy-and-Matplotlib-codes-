# **3. BAR PLOT**

```python
data={'joan':20,'will':22,'wesley':30,'william':35}
names=list(data.keys())
age=list(data.values())
fig=plt.figure(figsize=(10,10))
#creating a bar plot
plt.bar(names,age,color='magenta',width=0.5)
plt.xlabel("names")
plt.ylabel("age")
plt.title("barplot")
plt.plot()
```
